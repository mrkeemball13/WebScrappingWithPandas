# Data Retrieval

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Premier_League_clubs'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
# Only return the first table available in the url by using indexing -> [0]
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
soup.find_all('th')
```

```python
table.find_all('th')
```

```python
# Instead using " soup.find_all('th') ", I used " table.find_all('th') " where I only want to retrieve the 'th' in table variable
league_titles = table.find_all('th')
```

```python
league_titles
```

```python
#Only retrieve the title without the HTML format, .strip used to eliminate (\n) in the output
league_table_titles = [title.text.strip() for title in league_titles]
```

```python
print(league_table_titles)
```

```python
# Refine the name for each title
manual_corrections = {
    'Most recentpromotion': 'Most Recent Promotion',
    'Most recentrelegation': 'Most Recent Relegation',
    'Most recentfinish': 'Most Recent Finish',
    'Totalseasons': 'Total Seasons',
    'Totalspells': 'Total Spells',
    'Longestspell': 'Longest Spell',
    'Totalseasonsabsent': 'Total Seasons Absent',
    'Highestfinish': 'Highest Finish'

}

def correct_titles(titles, corrections):
    # Apply manual corrections to the titles
    corrected_titles = [corrections.get(title, title) for title in titles]
    return corrected_titles

# Correct the titles
league_table_titles = correct_titles(league_table_titles, manual_corrections)

print(league_table_titles)
```

# Data Transformation

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = league_table_titles)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
print(column_data)
```

```python

for row in column_data[1:]:
# [1:] -> is used for the "for loop" because the output without it will produce the blank row at row 0

    # take every data for each column and put it in one row
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    '''
    add the data to the dataframe by measure the length of the df,
    initially the df will have 0 length so, the first data will be inserted in index 0
    '''
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
import re

def correct_seasons_format(seasons):
    # Use regex to find patterns where a new year range starts without space
    corrected_seasons = re.sub(r'(\d{4}–\d{4})(?=\d{4})', r'\1 ', seasons)
    # Replace en dashes (–) with hyphens (-)
    corrected_seasons = corrected_seasons.replace('–', '-')
    return corrected_seasons

# Apply the correction to the "Seasons" column
df['Seasons'] = df['Seasons'].apply(correct_seasons_format)

```

```python
df
```

```python
def correct_most_recent_finish_format(finish):

    # Seperating the word and numbers, eg:League One16th -> League One 16th
    corrected_finish = re.sub(r'([a-zA-Z])(?=\d)|(\d)(?=[a-zA-Z])', r'\1\2 ', finish)

    return corrected_finish


# Apply the correction to the "Seasons" column
df['Most Recent Finish'] = df['Most Recent Finish'].apply(correct_most_recent_finish_format)
```
